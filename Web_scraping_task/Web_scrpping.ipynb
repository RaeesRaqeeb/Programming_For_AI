```python
from pdfminer.high_level import extract_text


def count_word_in_pdf(pdf_path):
    text = extract_text(pdf_path)
    
    words=text.split()
    
    print(words)
    word_count= len(words)
    total_char=0
    for i in words:
        total_char+=len(i)
    return word_count,total_char


pdf_path="Lab task 06 - Stack.pdf"

total_words,total_char=count_word_in_pdf(pdf_path)

print(f"Total number of words:",{total_words},"\ntotal characters:",{total_char})
```

```python
#Extracting text heading from the layout of the page
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextBox, LTTextLine, LTChar

def extract_heading():
    headings=[]
    for page_layout in extract_pages("Lab task 06 - Stack.pdf"):
        
        for element in page_layout:
            if(isinstance(element,LTTextBox)):
                for text_line in element:
                    if isinstance(text_line, LTTextLine):
                        for character in text_line:
                            if isinstance(character,LTChar):
                                if( character.size > 12):
                                    heading_text=text_line.get_text().strip()
                                    headings.append(heading_text)
                                    break
    return headings


print(extract_heading())
```
