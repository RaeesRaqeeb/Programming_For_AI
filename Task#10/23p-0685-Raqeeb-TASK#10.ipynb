```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

#Question #1

#Graph 1 

csv_file='Data.csv'

Data=pd.read_csv(csv_file)


#Plot1
longitude_x=[]
for i in Data['longitude']:
    longitude_x.append(float(i))

plt.figure(figsize=(12,12))
plt.subplot(3,3,1)
plt.hist(longitude_x,bins=50,color='blue',edgecolor='black',)
plt.grid(True)
plt.title('Total Rooms')


#plot 2

latitude_x=[]
for i in Data['latitude']:
    latitude_x.append((i))

plt.subplot(3,3,2)
plt.hist(latitude_x,bins=50,color='blue',edgecolor='black',)
plt.title("Latitude")
plt.grid(True)


#plot 3

housing_median_age_x=[]
for i in Data['housing_median_age']:
    housing_median_age_x.append(float(i))
    
plt.subplot(3,3,3)
plt.hist(housing_median_age_x,bins=50,color='blue',edgecolor='black',)
plt.title("Housing_median_age")
plt.grid(True)


#plot 4
total_rooms=[]
for i in Data['total_rooms']:
    total_rooms.append(float(i))
    
plt.subplot(3,3,4)
plt.hist(total_rooms,bins=50,color='blue',edgecolor='black',)
plt.title("Total ROoms")
plt.grid(True)

#Plot 5
total_bedrooms=[]
for i in Data['total_bedrooms']:
    total_bedrooms.append(float(i))
plt.subplot(3,3,5)

plt.hist(total_bedrooms,bins=50,color='blue',edgecolor='black',)
plt.title(" Total BedRooms")
plt.grid(True)

#Plot 6
population=[]
for i in Data['population']:
    population.append(float(i))

plt.subplot(3,3,6)
plt.hist(population,bins=50,color='blue',edgecolor='black',)
plt.title("Population")
plt.grid(True)


#Plot 7
households=[]
for i in Data['households']:
    households.append(float(i))

plt.subplot(3,3,7)
plt.hist(households,bins=50,color='blue',edgecolor='black',)
plt.title("Households")
plt.grid(True)

#Plot 8
median_income=[]
for i in Data['median_income']:
    median_income.append(float(i))

plt.subplot(3,3,8)
plt.hist(median_income,bins=50,color='blue',edgecolor='black',)
plt.title("Median Income")
plt.grid(True)

#Plot 9
median_house_value=[]
for i in Data['median_house_value']:
    median_house_value.append(float(i))

plt.subplot(3,3,9)
plt.hist(median_house_value,bins=50,color='blue',edgecolor='black',)
plt.title("Median House Value")
plt.grid(True)

```
