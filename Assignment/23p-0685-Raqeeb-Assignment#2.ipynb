```python
#Question 1
import os 
import xml.etree.ElementTree as ET
import csv

My_xml_file_path='/home/raqeeb/My_data/3rd_Semester/Programming_for_AI_Lab/Programmin_For_AI/Assignment/CTIDataset'
category_l=[]

Value_l=[]
print("WHat is in mytree?")
for file in os.listdir(My_xml_file_path):
    if file.endswith('.xml'):
        try:
            My_tree=ET.parse(os.path.join(My_xml_file_path,file))
            All_root=My_tree.getroot()
            Event_data=All_root.findall('Event')
            for  event in Event_data:
                for event_tags in event:
                    print(event_tags.text)
                    for items in event_tags:
                        for Cate in items:
                            if Cate.text is None:
                                        items.remove(Cate)
                            else:
                                    print(Cate.text)
                                    if(Cate.tag=='category'):
                                        category_l.append(Cate.text)
                                    if(Cate.tag=='value'):
                                        Value_l.append(Cate.text)
                                    
            My_tree.write(os.path.join(My_xml_file_path,file), encoding='utf-8', xml_declaration=True)

        except Exception as e:
            print(f"Error parsing {file}: {str(e)}")
            





print("*******************************************")
print("ALL NONE Type values are Removed from file")
print("*******************************************")


print("CATEGORIES",category_l)
print("*******************************************")

print("VALUES",Value_l)
print("*******************************************")

category_l.extend(Value_l)
# Create a list of data
data=[['Category','Value']]

for  i in range(len(category_l)):
    list1=[]
    list1.append(category_l[i])
    data.append(list1)

for i in range(len(Value_l)):
    data[i+1].append(Value_l[i])




print(data)
# Specify the CSV file path
csv_file = "MY_CSV_FILE.csv"

# Open the CSV file in write mode
with open(csv_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("Data written to", csv_file)
print("*******************************************")
```
