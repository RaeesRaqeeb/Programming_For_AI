```python
import pandas as pd

# Create first dataframe
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

# Create second dataframe
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [25, 30, 35, 40]
})

# Merge dataframes on 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print("Inner Merge:\n", merged_df)

# Merge dataframes with left join
left_join_df = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:\n", left_join_df)

# Merge dataframes with right join
right_join_df = pd.merge(df1, df2, on='ID', how='right')
print("\nRight Join:\n", right_join_df)

# Merge dataframes with outer join
outer_join_df = pd.merge(df1, df2, on='ID', how='outer')
print("\nOuter Join:\n", outer_join_df)
```

```python
#concatenation 
#first creating 3 data frames 

# DataFrame 1: Clothing Items
data1 = {
    'ItemID': [1, 2, 3],
    'ItemName': ['Abaya', 'Hijab', 'Jilbab'],
    'DesignerID': [101, 102, 101]
}

df_cloth1 = pd.DataFrame(data1)

# DataFrame 2: Designers
data2 = {
    'DesignerID': [101, 102],
    'DesignerName': ['Designer A', 'Designer B']
}

df_cloth2 = pd.DataFrame(data2)

# DataFrame 3: Clothing Items
data3 = {
    'ItemID': [4, 5],
    'ItemName': ['Kaftan', 'Tunic'],
    'DesignerID': [103, 104]
}

df_cloth3 = pd.DataFrame(data3)


print("DataFrame clothes 1:\n", df_cloth1)
print("\nDataFrame clothes 2:\n", df_cloth2)
print("\nDataFrame clothes 3:\n", df_cloth3)
```

```python
# Concatenate both DataFrames vertically
concat_df = pd.concat([df_cloth1, df_cloth3,df_cloth2], ignore_index=True)
print("\nConcatenated DataFrame:\n", concat_df)
# New row to be added
new_row = {'ItemID': 6, 'ItemName': 'Dress', 'DesignerID': 105, 'DesignerName': 'Designer C'}

# Convert the new row to a DataFrame
new_row_df = pd.DataFrame([new_row])

# Concatenate the new row to the existing DataFrame
concat_df = pd.concat([concat_df, new_row_df], ignore_index=True)
print("\nDataFrame after adding new row:\n", concat_df)
```

```python
# Concatenate with keys to create a MultiIndex
concat_multiindex = pd.concat([df_cloth1, df_cloth3], keys=['Modest_Clothing', 'Medieval_Clothing'])
print("\nConcatenated DataFrame with MultiIndex:\n", concat_multiindex)
# can we do multiindexing in single data frame ?
```

```python
import pandas as pd
import numpy as np

# Load the dataset
bank_data = pd.read_csv("bank_term_deposit.csv")

# 1) How many people are married, single and are divorced
marital_status_count = bank_data['marital'].value_counts()
print(marital_status_count)

# 2) Deviation of overall balance among the depositors
balance_sd = np.std(bank_data['balance'], ddof=1)
print(f"Standard Deviation of Balance: {balance_sd}")

# 3) Mean of the deposited balance
balance_mean = np.mean(bank_data['balance'])
print(f"Mean of Balance: {balance_mean}")

# 4) Average age
average_age = np.mean(bank_data['age'])
print(f"Average Age: {average_age}")

# 5) Display the Instances where depositors are in services but are married and single
services_married_single = bank_data[(bank_data['job'] == "services") & 
                                    ((bank_data['marital'] == "married") | 
                                     (bank_data['marital'] == "single"))]
print(services_married_single)
```
