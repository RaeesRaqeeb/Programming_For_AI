```python
#finding the outliners
#checking through graph
#Checking outliers for duration 

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.boxenplot(df_bank['duration'])


plt.subplot(1,2,2)
sns.histplot(df_bank['duration'])
```

__________________________________________________________________________________________________________________________________

```python
#Calculation Quartiles Q3 and Q1
Q1=df_bank['duration'].quantile(0.25)
Q3=df_bank['duration'].quantile(0.75)

IQR=Q3-Q1

Upper_limit= Q3 + 1.5*IQR
Lower_limit= Q1 - 1.5*IQR

Outliers_removed=df_bank[(df_bank['duration']>Lower_limit) & (df_bank['duration']<Upper_limit)]

```

<h3>Comparison of Dataset With and Without Outliers</h3>

```python
#Comparing with and without outliers
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
sns.boxenplot(df_bank['duration'])


plt.subplot(2,2,2)
sns.histplot(df_bank['duration'])

plt.subplot(2,2,3)
sns.boxenplot(Outliers_removed['duration'])

plt.subplot(2,2,4)
sns.histplot(Outliers_removed['duration'])
```
