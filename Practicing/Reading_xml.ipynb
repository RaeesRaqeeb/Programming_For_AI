```python
import os
import xml.etree.ElementTree as ET


xml_path='/home/raqeeb/My_data/3rd_Semester/Programming_for_AI_Lab/Programmin_For_AI/Practicing/CTIDataset'

for file1 in os.listdir(xml_path):
    if(file1.endswith('.xml')):
        file=ET.parse(os.path.join(xml_path,file1))
        root=file.getroot()
        events=root.findall('Event')
        for  event in events:
            print(event)
            for att in event:
                print(att)
                for att2 in att:
                    for item in att2:
                        if(item.tag=='category'):
                            att2.remove(item)
                        print(item.text)
        file.write(os.path.join(xml_path,file1), encoding='utf-8', xml_declaration=True)
            
```
