```python
#Question 2
import random
import statistics

financial_data_path="financial_data"

if not os.path.exists(financial_data_path):
    os.makedirs(financial_data_path)

My_dictionary={}

year=2022
day=1
month=1
Revenue=30000
Expenses=20000

My_dictionary['Date']=[]
My_dictionary['Revenue']=[]
My_dictionary['Expenses']=[]

l=0
while(l<=10):
    My_dictionary['Date'].append(f"{year}-{day}-{month}")
    My_dictionary['Revenue'].append(f"{Revenue}")
    My_dictionary['Expenses'].append(f"{Expenses}")
    day=random.randint(1,31)
    month=random.randint(1,13)
    Revenue=random.randrange(30000,90000,10000)
    Expenses=random.randrange(30000,80000,6500)
    l+=1

My_data_frame=pd.DataFrame(My_dictionary)
print(My_data_frame)
with open(os.path.join(financial_data_path,"financial_data.csv") , "w") as file:
    My_data_frame.to_csv(file,index=False)
    

#Profit for each month Revenue - Expense 
Revenue_list=[]
Revenue_list.extend(My_data_frame['Revenue'])


Expenses_list=[]
Expenses_list.extend(My_data_frame['Expenses'])
Profit_list=[]

for i in  range(len(My_data_frame['Date'])):
        Profit_list.append(int(Revenue_list[i]) - int(Expenses_list[i]))


#I Added another column of profit to the dataframe just to make it easy to understand
print("\nProfit column ADDED!!!!")
My_dictionary['Profit']=Profit_list
My_data_frame=pd.DataFrame(My_dictionary)
print(My_data_frame)
with open(os.path.join(financial_data_path,"financial_data.csv") , "w") as file:
    My_data_frame.to_csv(file,index=False)

total=0
num=0
for i in Profit_list:
    if(int(i)>0):
        total+=int(i)
        num+=1

average_profit=total/num

print("It is the average of Profit only not the loss which are negative values")
print("Average Profit:",average_profit)
```

```python
#Question #3

Student_data=np.random.randint(30,90,(3,3))

print(Student_data)

Avg_math=0
Avg_science=0
Avg_english=0

for i in Student_data:
        Avg_math+=i[0]
        Avg_science+=i[1]
        Avg_english+=i[2]
        
print("Math Average:",Avg_math/3)
print("English Average:",Avg_english/3)
print("Science Average:",Avg_science/3)

print("Average Score overall subject:",(Avg_science+Avg_english+Avg_math)/9)

#Top student
Student_1_avg=0
Student_2_avg=0
Student_3_avg=0

Student_1_avg+=(Student_data[0][0]+Student_data[0][1]+Student_data[0][2])
Student_2_avg+=(Student_data[1][0]+Student_data[1][1]+Student_data[1][2])
Student_3_avg+=(Student_data[2][0]+Student_data[2][1]+Student_data[2][2])
    

Comparing=[]
Comparing.extend([Student_1_avg/3,Student_2_avg/3,Student_3_avg/3])
print("Average Marks of Each student of all subject:",end=" ")
print(Comparing)
Maximum_average=max(Comparing)
print("Student:",Comparing.index(Maximum_average)+1," Scoured Maximum Average :",Maximum_average)

Subjects_list=[]
Subjects_list.extend([Avg_math,Avg_english,Avg_science])

Max_Subj_average_num=max(Subjects_list)

if(Subjects_list.index(Max_Subj_average_num) == 0):
    print("Student has improve highly in Math")
elif(Subjects_list.index(Max_Subj_average_num) == 1):
    print("Student has improve highly in English") 
elif(Subjects_list.index(Max_Subj_average_num) == 2):
    print("Student has improve highly in Science")
```
